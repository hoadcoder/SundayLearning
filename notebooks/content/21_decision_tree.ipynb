```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
```

```python
X, y = load_wine(return_X_y=True)
```

```python
load_wine()["feature_names"]
```

```python
dt_cls = DecisionTreeClassifier(criterion="entropy", min_samples_split=10, max_depth=2)

dt_cls = dt_cls.fit(X, y)
```

```python
plt.figure(figsize=(16, 8))
tree.plot_tree(dt_cls, filled=True);
```

##### Strenghts
- Fcil de entender e usar
- Permite voc interpretar dados facilmente e identificar regras
- Pode ser usada em algoritmos mais complexos como ensembles

##### Weaknesses
- So fceis de ter overfitting, principalmente quando voc tem dados com muitaaaaaaaaaaaaas features.
