### Matplotlib Components
- **Figure**: Janela, canvas ou página onde tudo será desenhado. 
- **Axes**: Area onde os dados são plotados com funções como plot() e scatter(). Pode ter ticks, labels, etc. associadas com isso. Isso explica porque figuras pode ter múltiplos axes. Dica: quando você vê plt.xlim, você está na verdade chamando ax.set_xlim() por baixo dos panos.
- **Axis** (Eixos): Eixos de um axes. Como eixo x e y

Não podemos ver, mas existem padrões que tomam conta da criação de alguns componentes, como a Figure e Axes.

Além disso, quando você vê, por exemplo um plt.xlim, o plt está chamando um ax.set_xlim() por baixo dos pontos. Na maioria dos casos usaremos as funções do plt por serem muito mais simples para gráficos simples. E buscaremos criar os axes e Figure manualmente para gráficos mais complexos.

Vejamos abaixo o que quero dizer com "limpo"

```python
iris = datasets.load_iris()
```

```python
with open("descricao_iris.txt", "w") as file:
    file.write(iris["DESCR"])
```

```python
iris_df = pd.DataFrame(iris["data"])
iris_df.columns = ["sepal_len", "sepal_wid", "petal_len", "petal_wid"]
iris_df.head()
```

```python
target = pd.DataFrame(iris["target"])
target.columns = ["output"]
target.head()
```

```python
df_iris = pd.concat([iris_df, target], axis=1)
```

```python
df_iris
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 16))
ax1, ax2, ax3, ax4 = ax.flat

sns.heatmap(df_iris[["sepal_len", "sepal_wid", "petal_len", "petal_wid"]], ax=ax2)

ax2.set_ylabel("Índices")
ax2.set_xlabel("Atributos");
```

```python
def normalizer(vetor):
    new_vetor = []
    for xi in vetor:
        zi = (xi - np.min(vetor)) / (np.max(vetor) - np.min(vetor))
        new_vetor.append(zi)
    
    return new_vetor
```

```python
df_iris["norm_sepal_len"] = normalizer(df_iris["sepal_len"])
df_iris["norm_sepal_wid"] = normalizer(df_iris["sepal_wid"])
df_iris["norm_petal_len"] = normalizer(df_iris["petal_len"])
df_iris["norm_petal_wid"] = normalizer(df_iris["petal_wid"])
```
